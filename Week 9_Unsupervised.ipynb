# Unsupervised Learning Use Cases

This notebook demonstrates common use cases of unsupervised learning, including clustering and dimensionality reduction, with practical examples and explanations.

## 1. Clustering with K-Means

Clustering is a technique to group similar data points together. K-Means is a popular clustering algorithm that partitions data into K distinct clusters based on feature similarity.

```python
# Example: K-Means Clustering on Iris Dataset
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

# Visualize the clusters using the first two features
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', label='Cluster')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering on Iris Dataset')
plt.show()
```

### Explanation

- **K-Means** groups the iris data into 3 clusters based on feature similarity.
- The plot shows how the algorithm separates the data points into clusters.
- No labels are used during training; the algorithm finds structure in the data on its own.

## 2. Dimensionality Reduction with PCA

Dimensionality reduction simplifies data by reducing the number of features, making it easier to visualize and analyze. Principal Component Analysis (PCA) is a widely used technique for this purpose.

```python
# Example: PCA on Iris Dataset
from sklearn.decomposition import PCA

# Reduce to 2 principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualize the reduced data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target, cmap='Set1', label='Species')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Iris Dataset')
plt.show()
```

### Explanation

- **PCA** reduces the 4-dimensional iris data to 2 dimensions.
- The plot shows how the main variance in the data can be captured in fewer dimensions.
- This helps in visualizing and understanding the structure of high-dimensional data.

## 3. Anomaly Detection with Isolation Forest

Anomaly detection identifies rare items or events that differ significantly from the majority of the data. Isolation Forest is an unsupervised algorithm for detecting anomalies.

```python
# Example: Anomaly Detection on Synthetic Data
from sklearn.ensemble import IsolationForest
import numpy as np

# Generate synthetic data
rng = np.random.RandomState(42)
X_normal = 0.3 * rng.randn(100, 2)
X_outliers = rng.uniform(low=-4, high=4, size=(20, 2))
X_anomaly = np.r_[X_normal + 2, X_normal - 2, X_outliers]

# Fit Isolation Forest
iso_forest = IsolationForest(contamination=0.1, random_state=42)
y_pred = iso_forest.fit_predict(X_anomaly)

# Visualize anomalies
plt.scatter(X_anomaly[:, 0], X_anomaly[:, 1], c=y_pred, cmap='coolwarm', label='Anomaly')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Anomaly Detection with Isolation Forest')
plt.show()
```

### Explanation

- **Isolation Forest** separates normal data from anomalies without using labels.
- Points predicted as anomalies are shown in a different color.
- Useful for fraud detection, fault detection, and more.

## Summary

Unsupervised learning helps uncover hidden patterns in data without labeled outcomes. Common use cases include clustering, dimensionality reduction, and anomaly detection.
